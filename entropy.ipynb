```julia
include("./src/OTOC.jl")
include("./src/hamiltonians.jl")
using Convex
using Plots
```

```julia
nbit = 10
nfront = 5
nnext = nbit - nfront
h = xxz(nbit,1,1,15)
quench = xxz(nbit,1,1,15) + 6*put(nbit,1=>Z);
hmat = Matrix(mat(h))
states = eigvecs(hmat);
state = states[:,32];
tlist = 10 .^ range(-1,2,100);
```

```julia
von_en = []
for _ in 1:1000
    finalstates = [UMatrix(t,quench) * state for t in tlist];
    rholist = [finalstate * finalstate' for finalstate in finalstates];
    von_s = [partialtrace(r,1,[2^nfront,2^nnext]) * log(partialtrace(r,1,[2^nfront,2^nnext])) for r in rholist];
    entropy = -tr.(von_s);
    push!(von_en,entropy)
end
```

```julia
von_en = real(von_en)
```

```julia
vs = [sum(reduce(vcat,transpose.(von_en))'[i,:]) / 1000 for i in 1:100]
```

```julia
scatter(tlist,vs,xscale=:log10)
```
