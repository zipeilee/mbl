```julia
include("OTOC.jl")
using Plots
```

```julia
function rand_sz(nbit::Int,h,periodic::Bool=false)
    sx = i ->  1/2 * put(nbit, i=>X)
    sy = i ->  1/2 * put(nbit, i=>Y)
    sz = i ->  1/2 * put(nbit, i=>Z)
    interaction = map(1: ( periodic ? nbit : nbit-1) ) do i
        j = i%nbit + 1
        (sz(i)*sz(j))
    end |> sum

    hi = rand(-h:.0001:h,nbit)
    disorder = map(1:nbit) do i 
        hi[i] * sz(i)
    end |> sum

    interaction + disorder
end
```

```julia
h0 = rand_sz(8,0);
h1 = rand_sz(8,5);
```

```julia
tlist = range(0,4,100)
y0 = real([sum([inf_otoc(X,X; t=t,hamiltonian=h0,nbit=8, i=8,j=2) for _ in 1:1000])/1000 for t in tlist]);
y1 = real([sum([inf_otoc(X,X; t=t,hamiltonian=h1,nbit=8, i=8,j=2) for _ in 1:1000])/1000            for t in tlist]);
```
